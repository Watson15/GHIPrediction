```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = 'Datasets/stationsName_lat_long_data.csv'
numAuxStations = 4
stationsName_lat_long_datadf = pd.read_csv(file_path, delimiter=',', on_bad_lines='skip')
for row in stationsName_lat_long_datadf.itertuples():
    latitude = row.Latitude
    longitude = row.Longitude
    station = row.station
    print(latitude, longitude, station)
```

```python
def plot_stations_matplotlib(all_stations_df, figsize=(8, 6)):

    plt.figure(figsize=figsize)

    # Plot unpicked stations (from all_stations_df, excluding the neighbors and center)
    for _, row in all_stations_df.iterrows():
        plt.scatter(row['Longitude'], row['Latitude'], color='gray', alpha=0.9, label='_nolegend_')
            #plt.text(row['Longitude'] + 0.002, row['Latitude'], row['station'], fontsize=8, color='gray')
    min_lon, max_lon = all_stations_df['Longitude'].min() - 0.5, all_stations_df['Longitude'].max() + 0.5
    x = np.linspace(min_lon, max_lon, 300)
    m = -1.1
    b = -88
    y = m * x + b
    #plt.plot(x, y, color='green', linewidth=2)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Target Station and Spatially Diverse Neighbors')
    plt.legend()
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='datalim')
    plt.show()
plot_stations_matplotlib(stationsName_lat_long_datadf)
```
